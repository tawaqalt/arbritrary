## **Problem 1**

```python
#creating random numbers
import numpy as np
mean = [-3,0]
cov = [[1.0,0.8],[0.8,1.0]]
np.random.seed(0)
rand_num = np.random.multivariate_normal(mean,cov,500)
print(rand_num)
print(rand_num.shape)
```

## **Problem 2**

```python
#Visualization by scatter plot
import matplotlib.pyplot as plt
plt.scatter(rand_num[:,0],rand_num[:,1])
plt.title("Visualization of Scatter Plot")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

## **Problem 3**

```python
#Visualization by histogram x1
plt.title("Visualization of Histogram")
plt.xlabel("X1")
plt.ylabel("Frequency")
plt.xlim(-5)
plt.hist(rand_num[:,0])
plt.show()
```

```python
#Visualization by Histogram x2
plt.title("Visualization of Histogram")
plt.xlabel("X2")
plt.ylabel("Frequency")
plt.xlim(-5)
plt.hist(rand_num[:,1])
plt.show()
```

## **Problem 4**

```python
#Create 500 new random numbers with a two-dimensional normal distribution whose mean is (0, -3)
import numpy as np
mean = [0, -3]
cov = [[1.0,0.8],[0.8,1.0]]
np.random.seed(0)
new_rand_num = np.random.multivariate_normal(mean,cov,500)
print(new_rand_num)
print(new_rand_num.shape)
```

```python
#Visualizing by scatter plot Problem 1 and 4 given 0 for problem 1 and 1 for problem 4
import matplotlib.pyplot as plt
plt.title("Visualization of Scatter Plot")
plt.xlabel("X1")
plt.ylabel("X2")
plt.scatter(rand_num[:,0],rand_num[:,1])
plt.scatter(new_rand_num[:,0],new_rand_num[:,1])
plt.legend(["0","1"])
plt.show()
```

# **Problem 5**

```python
#Concatenating the two arrays
new_array = np.concatenate([rand_num,new_rand_num],axis=0)
print(new_array)
print(new_array.shape)
```

## **Problem 6**

```python
# for problem 1, the result is 500, 2 using append
zeros = np.zeros(500)
ones = np.ones(500)
new_column = np.append(zeros,ones)
new_ndarr = np.append(new_array,new_column[:,None],1)
print(new_ndarr)
print(new_ndarr.shape)
```
