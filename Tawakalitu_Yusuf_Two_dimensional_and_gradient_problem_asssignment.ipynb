## **Problem 1**

```python
#Creating an array for both x and y in the interval -50 to 50
import numpy as np
x = np.arange(-50, 50.1, 0.1)
y = x/2 + 1
print(x)
print(y)
print(x.shape)
```

# **Problem 2**

```python
z_array = np.concatenate((x[:, np.newaxis], y[:, np.newaxis]), axis=1)
print(z_array)
print(z_array.shape)
```

## **Problem 3**

```python
#the gradient at each x-coordinate
the_gradient = (z_array[1:, 1] - z_array[:-1, 1]) / (z_array[1:, 0] - z_array[:-1, 0])
print(the_gradient)
print(the_gradient.shape)
```

## **Problem 4**

# **Problem 5**

```python
#Python functionalization
def compute_gradient(function, x_range =(-50, 50.1, 0.1)):
  array_x = np.arange(-50, 50.1, 0.1)
  array_y = function(array_x)
  z_array = np.concatenate((array_x[:, np.newaxis], array_y[:, np.newaxis]), axis=1)
  the_gradient = (z_array[1:, 1] - z_array[:-1, 1]) / (z_array[1:, 0] - z_array[:-1, 0])
  return z_array, the_gradient

def function1(array_x):
     y_array = array_x ** 2
     return y_array

z_array1, the_gradient1 = compute_gradient(function1)
print(compute_gradient(function1))

```

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    array_x = np.arange(*x_range)
    array_y = function(array_x)
    z_array = np.column_stack((array_x, array_y))
    the_gradient = (z_array[1:, 1] - z_array[:-1, 1]) / (z_array[1:, 0] - z_array[:-1, 0])
    return z_array, the_gradient

def function3(array_x):
    y_array = np.sin(array_x**0.5)
    return y_array

# Call the compute_gradient function
z_array3, the_gradient3 = compute_gradient(function3)

print(compute_gradient(function3))


```

```python
# Plot the graph
plt.plot(z_array3[:, 0], z_array3[:, 1], label='Function 3')
plt.title('Function 3 and its Gradient')
plt.xlabel('x')
plt.ylabel('y')
plt.show()


plt.title('Function 1 and its Gradient')
plt.xlabel('x')
plt.ylabel('Gradient')
plt.plot(z_array3[:-1,0], the_gradient3)
plt.show()
```
