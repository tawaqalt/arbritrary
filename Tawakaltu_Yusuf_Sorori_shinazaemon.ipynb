# **Problem 1: Rice grains**

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_rice_grains(day):
    return 2**(day - 1)

number_of_rice = np.array([calculate_rice_grains(day) for day in range(1, 101)])

print("\nNumber of rice grains at the 100th day:")
print(np.sum(number_of_rice))
plt.xlabel("Number of Day")
plt.ylabel("Number of rice grains")
plt.title("Number of rice grains from 1st- 100th day")
plt.plot(number_of_rice, label="rice grains",linewidth=0.5,linestyle="-", color="blue")
plt.show()
```

```python
import matplotlib.pyplot as plt

def compute_sorori_shinzaemon(n_days=100):
    """
    A function that calculates the number of rice grains per day in the rice anecdote of Shinzaemon Sorori.

    Parameters
    ----------
    n_days: int
        Number of days to get rice (default : 100)

    Returns
    -------
    list_n_grains : list
        List of the number of rice you get one day (list length equals n_days)
    list_total_grains : list
        List of total number of rice received by a certain day (list length is equal to n_days)
    """

    list_n_grains = [2**i for i in range(n_days)]
    list_total_grains = [sum(list_n_grains[:i+1]) for i in range(n_days)]

    return list_n_grains, list_total_grains

def plot_grains(n_days):
    list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days)

    plt.figure(figsize=(8, 6))
    plt.plot(list_n_grains, label='Grains received each day')
    plt.plot(list_total_grains, label='Total grains received')
    plt.title('Number of Rice Grains in Sorori Shinzaemon Anecdote')
    plt.xlabel('Days')
    plt.ylabel('Number of Rice Grains')
    plt.legend()
    plt.show()

plot_grains(360)
```

```python
def calculate_days(n_grains, n_people):
    """
    A function that calculates how many days a given number of people can live on a given number of grains of rice.

    Parameters
    ----------
    n_grains: int
        Number of grains of rice
    n_people: int
        Number of people

    Returns
    -------
    n_days: int
        Number of days the people can live on the rice
    """
#Based on my research, an adult typically consumes about 1/2 cup of uncooked rice per meal, which is approximately 93 grams.
#Whole grain rice is often recommended, and adults should aim for at least 3-4 ounces (85-113 grams) of whole grain rice per day2.
#Lets assume that each person in the class eats about 100 grams of uncooked rice per day.
#Given that there are about 48,000 grains of rice in a kilogram, this equates to about 4800 grains of rice per day per person.

    grains_per_person_per_day = 4800
    total_grains_needed_per_day = n_people * grains_per_person_per_day

    n_days = n_grains // total_grains_needed_per_day

    return n_days


calculate_days(1267650600228229401496703205375, 60)
```
