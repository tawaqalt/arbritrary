Scratch Implementation

```python
import numpy as np

class ScratchLinearRegression:
    """
    Scratch implementation of linear regression

    Parameters
    ----------
    num_iter : int
        Number of iterations
    lr : float
        Learning rate
    no_bias : bool
        True if no bias term is included
    verbose : bool
        True to output the learning process

    Attributes
    ----------
    self.theta : ndarray, shape (n_features + 1,)
        Parameters
    self.loss : ndarray, shape (self.num_iter,)
        Record losses on training data
    self.val_loss : ndarray, shape (self.num_iter,)
        Record loss on validation data
    """

    def __init__(self, num_iter=1000, lr=0.01, no_bias=False, verbose=False):
        self.num_iter = num_iter
        self.lr = lr
        self.verbose = verbose
        self.no_bias = no_bias
        self.theta = None
        self.loss = np.zeros(self.num_iter)  # Initialize self.loss
        self.val_loss = np.zeros(self.num_iter)  # Initialize self.val_loss

    def fit(self, X, y, X_val=None, y_val=None):
        if not self.no_bias:
            X = np.concatenate([np.ones((X.shape[0], 1)), X], axis=1)

        n_samples, n_features = X.shape
        self.theta = np.zeros(n_features)

        for i in range(self.num_iter):
            predictions = X.dot(self.theta)
            errors = predictions - y
            gradients = X.T.dot(errors) / n_samples
            self.theta -= self.lr * gradients

            # Calculate and store the loss for training and validation data
            self.loss[i] = np.mean((predictions - y) ** 2)

            if X_val is not None and y_val is not None:
                val_predictions = self.predict(X_val)
                self.val_loss[i] = np.mean((val_predictions - y_val) ** 2)

            if self.verbose:
                print(f"Iteration {i + 1}: Loss = {self.loss[i]:.4f}", end="")
                if X_val is not None and y_val is not None:
                    print(f", Validation Loss = {self.val_loss[i]:.4f}")
                else:
                    print()

    def predict(self, X):
        """
        Estimate using linear regression.

        Parameters
        ----------
        X : ndarray, shape (n_samples, n_features)
            Sample

        Returns
        -------
        ndarray, shape (n_samples,)
            Estimated result by linear regression
        """
        n_samples = X.shape[0]
        if not self.no_bias:
            X = np.concatenate([np.ones((n_samples, 1)), X], axis=1)
        return X.dot(self.theta)
```

# 【Problem 4】Mean squared error

```python
def MSE(y_pred, y):
    """
    Calculation of mean square error

    Parameters
    ----------
    y_pred : of the following form. ndarray, shape (n_samples,)
      Estimated value
    y : of the following form. ndarray, shape (n_samples,)
      Correct answer value

    Returns
    ----------
    mse : numpy.float
      Mean squared error
    """
    mse = np.mean((y_pred - y) ** 2)
    return mse
```

```python
y_test = np.array([2,3,1,4,5])
y_pred = np.array([1,3,3,2,2])
mse = MSE(y_pred, y_test)
print(mse)
```

# 【Problem 7】Plotting the learning curve

```python
# Scaling the input feature
X_train = sc.fit_transform(X_train.reshape(-1, 1))
X_test = sc.transform(X_test.reshape(-1, 1))

# Split the training data into train and validation
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Create an instance of ScratchLinearRegression with a smaller learning rate and more iterations
instance_lr = ScratchLinearRegression(num_iter=5000, lr=0.001, no_bias=False, verbose=False)

# Fit the model with validation data
instance_lr.fit(X_train, y_train, X_val, y_val)

# Plotting the learning curve
fig, ax = plt.subplots(figsize=(10, 5))
plt.plot(range(len(instance_lr.loss)), instance_lr.loss, color='blue', label='train')
plt.plot(range(len(instance_lr.val_loss)), instance_lr.val_loss, color='red', label='validation')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
