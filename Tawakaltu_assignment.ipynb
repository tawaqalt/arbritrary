## **Problem 3**

```python
THICKNESS = 0.00008
folded_thickness = THICKNESS

for _ in range(43):
    folded_thickness = folded_thickness * 2

print("Thickness of paper when folded 43 times: {:.2f} meters".format(folded_thickness))
print("Thickness of paper in kilometres is: {: .2f} km".format(folded_thickness/1000))
```

## **Problem 6**

```python
THICKNESS = 0.00008

# Create an empty list to store thickness values
thickness_values = []

# Add the initial thickness to the list
thickness_values.append(THICKNESS)

# Calculate and store the thickness after each fold in a loop
for _ in range(43):
    THICKNESS *= 2  # Double the thickness with each fold
    thickness_values.append(THICKNESS)

# Verify that 44 values are stored in the list
print("Number of values in the list:", len(thickness_values))
```

'the thickness of the paper was at a constant value and only increased at 35 folds and a significant increase was seen at 40+folds. This graph just shows how an exponential function works.'
