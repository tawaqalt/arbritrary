## **Problem 5**

Skewness measures the asymmetry of the probability distribution of a real-valued random variable. In simpler terms, it indicates whether a dataset is skewed to the left (negatively skewed), skewed to the right (positively skewed), or approximately symmetric.

Positive Skewness: The tail on the right side is longer or fatter than the left side. The mass of the distribution is concentrated on the left side, and the right tail is longer.
Negative Skewness: The tail on the left side is longer or fatter than the right side. The mass of the distribution is concentrated on the right side, and the left tail is longer.
Skewness = 0: The distribution is perfectly symmetric.
Skewness is often used to identify the direction and degree of asymmetry in a dataset.

Kurtosis:
Kurtosis measures the "tailedness" of the probability distribution of a real-valued random variable. It provides information about the heaviness of the tails in relation to a normal distribution.

Leptokurtic (Positive Kurtosis): Tails of the distribution are heavier than the normal distribution. The peak of the distribution is sharper.
Mesokurtic (Zero Kurtosis): The distribution has tails of similar weight as a normal distribution.
Platykurtic (Negative Kurtosis): Tails of the distribution are lighter than the normal distribution. The peak of the distribution is flatter.
In essence, kurtosis measures the presence of outliers or extreme values in a dataset.

Interpretation:
High skewness or kurtosis values suggest departures from normality.
Understanding skewness and kurtosis is important in statistical analysis, especially in finance, risk management, and data exploration.

````python
import seaborn as sns
"--- objective varaible is the same as target variable ---"
a = dataset_housing["SalePrice"].skew()
b = dataset_housing["SalePrice"].kurt()
print("the skewness = {:.2f}    the kurtosis = {:.2f} ".format(a, b))
sns.distplot(target_variable, kde = False, rug = False, bins = 10)
````

````python
# performing a logarithmic transformation on the objective variable
import numpy as np

log_tgv = np.log(target_variable)
log_tgv
````

````python
#Distribution of the logarithmic transformation
print("the skewness = {:.2f}    the kurtosis = {:.2f} ".format(log_tgv["SalePrice"].skew(), log_tgv["SalePrice"].kurt()))

sns.distplot(log_tgv, kde = False, rug = False, bins = 10)
````

````python
#the histplot
sns.histplot(log_tgv, kde = False, bins = 10)
````

* the difference between the logarithmic transformation and the original objective variable is that the logarithmic transformation has a positive skewness and a negative kurtosis.hence the distribution of the logarithmic transformation is more normal than the original objective variable.
